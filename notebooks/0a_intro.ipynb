## Importing packages

By convention, imports go at the top of a Python script or notebook (see [PEP 8](https://peps.python.org/pep-0008/#imports)). In relevant part:

> Imports are always put at the top of the file, just after any module comments and docstrings, and before module globals and constants.

> Imports should be grouped in the following order:

> - Standard library imports.
> - Related third party imports.
> - Local application/library specific imports.

> You should put a blank line between each group of imports.

For this notebook, we will use a few packages that are already installed in the course environment.

`````python
# standard library
import sys
import time

# third-party packages
import plotly.express as px
import polars as pl
`````

Note a few things in the block above.

1. `import sys` is the simplest version.
1. For some things we use a lot, and also by convention, we abbreviate the names of some packages. For example, `pandas` is often imported as `pd`.
1. `polars` is often imported as `pl`, and we will use Polars heavily on Day 1.
1. The lines that start with `#` are comments. Those lines are not executed by Python, and they are useful for us to make notes about what we are doing.

Let's see how these work in action.

`````python
print(sys.executable)
`````

Note that to find the contents of this attribute `executable` within the `sys` package, we have to use the package namespace `sys`. More about namespaces below.

## Namespaces

Somewhat abstractly, the [Python docs](https://docs.python.org/3/tutorial/classes.html#python-scopes-and-namespaces) define a namespace as follows.

> A namespace is a mapping from names to objects.

For our purposes, we can think of namespaces as paths to get to tools of interest. This topic goes much deeper, but a practical understanding is fine for our use.

If we want to know what is contained in a namespace, we can easily find out with the `dir()` built-in function.

`````python
dir(sys)
`````

While we might think of namespaces as synonymous with packages, the idea is more general than that.
Individual objects and modules have their own namespaces.

Let's inspect the Polars namespace, too.

`````python
dir(pl)
`````

We can also look at what is in the global namespace for this notebook.

`````python
dir()
`````

Note that we see the `sys`, `time`, `pd`, `px`, and `pl` names that we imported earlier.

## Rich text

Jupyter uses the simple [Markdown](https://daringfireball.net/projects/markdown/) syntax for formatting text. There are some extensions and differences from original Markdown, so you may find the [Jupyter Notebook docs](https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Working%20With%20Markdown%20Cells.html) to be a better reference.

- **Bold** a word by enclosing it in pairs of asterisks: `**Bold**`.
- *Italicize* a word by enclosing it in single asterisks: `*Italicize*`.
- ***Do both*** with three asterisks: `***Do both***`.
- We can also use headings by starting the line with one or more pound signs, where one is a top-level heading: `#`.

# First heading

## Second heading

### Third heading

```text
# First heading
## Second heading
### Third heading
```

# Notebook challenges

There are a few challenges when using notebooks.

1. The cell structure is flexible, but it will allow you to do things out of order, and anything you import or assign is still there, even if you change the code. To avoid this issue:
    - Purposefully try to keep your code in the order that it should be run.
    - Periodically, restart the kernel and run all cells to make sure that your notebook runs in order.
1. If you want to version control projects, you might remove cell outputs before committing, like the course repository does, to keep comparisons readable.
1. It is not a good format for every kind of final research output. For polished documents, slides, manuscripts, and websites, Quarto is a better fit.

Overall, the notebook is a great tool.
Once you have some experience using notebooks, you may find them fairly natural to work with.
